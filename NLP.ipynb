I have used the NLTK library for natural language processing tasks.
I have downloaded an ebook "The Tales of Chinatown" from the provided link and made a function as mentioned in the instruction.

```python
path = '/New Text Document.txt'      #The path of the text file
```

```python
read_file = open(path,'r')            #Reading the text file
chinatown = read_file.read()
```

```python
import nltk
nltk.download('punkt')

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk import sent_tokenize
import re
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

The function completes following tasks:

1. Perform word tokenization (split the text into words).

2. Convert all tokenized words to lower case.

3. Remove stopwords and punctuation.

4. Implement stemming or lemmatization to identify words with the same root.

5. Count each of the unique words in your dataset.

```python
REPLACE = re.compile('[/(){}\[\]\|@,;”“,]')
SYMBOLS = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))

def language_processing(text_file):
  nltk_tokens = nltk.word_tokenize(text_file)   #Tokenize words
  text_file = text_file.lower() # lowercase text
  text_file = REPLACE.sub(' ', text_file) # replace REPLACE symbols by space in text
  text_file = SYMBOLS.sub('', text_file) # delete symbols which are in SYMBOLS from text
  text_file = ' '.join(word for word in text_file.split() if word not in STOPWORDS) # delete stopwors from text
  new = text_file
  nltk_tokens = nltk.word_tokenize(new)   #Generating new tokens
  print("Stemming for 'The Tales of Chinatown'")
  porter_stemmer = PorterStemmer()                #Implementing the Stemming for identifying the same roots
  for i in nltk_tokens:
    print ("Actual: %s  Stem: %s" % (i,porter_stemmer.stem(i)))
  print("Lemmatization for 'The Tales of Chinatown'")
  wordnet_lemmatizer = WordNetLemmatizer()             #Implementing the Lemmatization for identifying the same roots
  for j in nltk_tokens:
    print ("Actual: %s  Lemma: %s"  % (j,wordnet_lemmatizer.lemmatize(j)))
  fdist1 = nltk.FreqDist(nltk_tokens)                    #Counting the Unique Words in the text
  w_freq = dict((word, freq) for word, freq in fdist1.items() if not word.isdigit())
  print(w_freq)



#EXECUTION OF THE FUNCTION
language_processing(chinatown)
```
